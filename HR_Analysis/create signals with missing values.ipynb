```python
for file in files_dist:
    signal_ex=pd.read_pickle('./signals/died/'+file)
    if len(signal_ex) == 0:
        continue
    index=pd.MultiIndex.from_product( iterables=[signal_ex.type.unique(),signal_ex.complex_id.unique()],names=['type','complex_id'])
    signal_ex_ind=signal_ex.reset_index(drop=True)
    signal_ex_ind=signal_ex_ind.set_index(['type','complex_id'])
    signal_reindexed=signal_ex_ind.reindex(index)
    R=signal_reindexed.loc['R',:]
    RR=signal_reindexed.loc['R',:].diff()
    Q=signal_reindexed.loc['Q',:]
    S=signal_reindexed.loc['S',:]
    T=signal_reindexed.loc['T',:]
    df_complete=pd.concat([R,RR,Q,S,T],axis=1)
    df_complete.columns=['R','R_val','RR','RR_val','Q','Q_val','S','S_val','T','T_val']
    df_complete.to_pickle('./signals/died/with_missing/'+file)
    print(file)
```
